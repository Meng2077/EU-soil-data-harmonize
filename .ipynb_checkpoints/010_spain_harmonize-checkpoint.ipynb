### Parcelas INES

```python
import pandas as pd
import numpy as np

# read in data
data_path = '/mnt/inca/ai4sh_data.harmo'
df1 = pd.read_excel(f'{data_path}/raw_data/Spain/bbdd-cos/1_BDCOS10_22545_ParcelasINES/COS10_22545_ParcelasINES_Metadatos.xlsx', sheet_name='COS10_22545_ParcelasINES')

df1 = df1.rename(columns={'Arcilla':'clay','DAcorr':'bulk_density','C%':'oc'})
df1['oc'] = df1['oc']*13 # % -> g/kg, walkey-black method

```

```python
from pyproj import Proj, transform
df1 = df1[df1['huso']!=0]
proj_utm = {zone: Proj(proj='utm', zone=zone, ellps='WGS84', datum='WGS84') for zone in df1['huso'].unique()}
proj_latlon = Proj(proj='latlong', datum='WGS84')

def convert_utm_to_latlon(row):
#     if row['huso'] == 0:
#         return pd.Series([np.nan,np.nan])
#     utm_proj = proj_utm[row['huso']]
    utm_proj = proj_utm[30]
    lon, lat = transform(utm_proj, proj_latlon, row['XUTM_ETRS'], row['YUTM_ETRS'])
    return pd.Series([lat, lon])

# Apply the conversion for each row
df1[['lat', 'lon']] = df1.apply(convert_utm_to_latlon, axis=1)
```

```python
# compile the data with provinces and corresponding year
import geopandas as gpd
spain = gpd.read_file('/mnt/inca/tillage_index/data/EUROSTATS/nuts2/NUTS_RG_20M_2021_3035.shp')
spain = spain.loc[spain['CNTR_CODE']=='ES']
spain = spain.loc[spain['LEVL_CODE']==3]
len(spain)


tl = []
with open('/mnt/inca/ai4sh_data.harmo/raw_data/Spain/province_year.txt', 'r') as file:
    lines_list = file.readlines()

tl = [line.strip() for line in lines_list]
province_year_dict = dict(zip(tl[0:50],tl[-50::]))

spain['year'] = 0
for index,row in spain.iterrows():
    spain.at[index, 'year'] = province_year_dict.get(row['NUTS_NAME'])
#     if row['NUTS_NAME'] in province_year_dict:
#         del province_year_dict[row['NUTS_NAME']]
    
spain.loc[spain['NUTS_NAME']=='Valencia/València','year'] = 2009
spain.loc[spain['NUTS_NAME'].isin(['Tenerife','La Gomera','El Hierro','La Palma']),'year'] = 2005 # 'Santa Cruz de Tenerife': '2005'
spain.loc[spain['NUTS_NAME']=='Ourense','year'] = 2003
spain.loc[spain['NUTS_NAME']=='León','year'] = 2010
spain.loc[spain['NUTS_NAME'].isin(['Fuerteventura','Gran Canaria','Lanzarote']),'year'] = 2005 # 'Las Palmas de Gran Canaria': '2005'
spain.loc[spain['NUTS_NAME']=='Castellón/Castelló','year'] = 2009
spain.loc[spain['NUTS_NAME']=='Ávila','year'] = 2012
spain.loc[spain['NUTS_NAME']=='Araba/Álava','year'] = 2018
spain.loc[spain['NUTS_NAME']=='Alicante/Alacant','year'] = 2006
spain.loc[spain['NUTS_NAME'].isin(['Menorca','Mallorca','Eivissa y Formentera']), 'year'] = 2003 # 'Baleares': '2003'

spain.crs = 'EPSG:3035'

spain = spain.to_crs('EPSG:4326')
```

```python
from shapely.geometry import Point

df1 = df1.loc[df1['lat'].notna()]
df1 = df1.loc[df1['lon'].notna()]

gdf_points = gpd.GeoDataFrame(df1, geometry=[Point(xy) for xy in zip(df1.lon, df1.lat)])
gdf_points.crs = "EPSG:4326"

gdf_joined = gpd.sjoin(gdf_points, spain, how="left", op='within')

edge = gdf_points.loc[gdf_joined['year'].isna()]
edge = gpd.GeoDataFrame(edge) # , geometry=[Point(xy) for xy in zip(edge.lon, edge.lat)]
gdf = gdf_joined.loc[gdf_joined['year'].notna()]
gdf = gpd.GeoDataFrame(gdf)
edge.crs = "EPSG:4326"
gdf.crs = "EPSG:4326"

edge = edge.to_crs("EPSG:25830")  # Convert to local UTM zone 30N
edge['geometry'] = edge.buffer(6000) # buffer in meters
edge = edge.to_crs("EPSG:4326")

edge_joined = gpd.sjoin(edge, spain, how="left", predicate='intersects')
edge_joined.loc[edge_joined['year'].isna(),'year'] = 2003

gdf = pd.concat([gdf, edge_joined], ignore_index=True)
gdf['hzn_top'] = 0
gdf['hzn_btm'] = 10
```

```python
temp = gdf[['hzn_top','hzn_btm','year','lat','lon','bulk_density','clay','oc']]
temp = temp.rename(columns={'year':'time'})
temp['ref'] = 'spain.ParcelasINES'
temp['nuts0'] = 'ES'
temp['point_id'] = gdf['CodParcela']
temp.to_csv(f'{data_path}/data/spain.ParcelasINES_harmonized_v1.csv',index=False)

temp.head(5)
```

```python

data = temp
print(len(edge))

import matplotlib.colors as mcolors
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11, 8))
hexbin = ax.hexbin(data['lon'], data['lat'], gridsize=50, cmap='RdYlGn', mincnt=1) #, norm=mcolors.LogNorm()
ax.set_xlabel('Longitude', fontsize=14)
ax.set_ylabel('Latitude', fontsize=14)
ax.set_title(f'data distribution', fontsize=16)

# ax.set_ylim([36,44])
# ax.set_xlim([-10,5])
colorbar = plt.colorbar(hexbin)
colorbar.set_label(f'count', fontsize=14)

plt.grid(True)
plt.show()
```

### Parcelas COS

```python
import pandas as pd
import numpy as np
data_path = '/mnt/inca/ai4sh_data.harmo'

df21 = pd.read_excel(f'{data_path}/raw_data/Spain/bbdd-cos/2_BDCOS1030_800_ParcelasCOS/COS1030_800_ParcelasCOS_Metadatos.xlsx', sheet_name='ParcelaCOS1030_1')
df22 = pd.read_excel(f'{data_path}/raw_data/Spain/bbdd-cos/2_BDCOS1030_800_ParcelasCOS/COS1030_800_ParcelasCOS_Metadatos.xlsx', sheet_name='ParcelasCOS1030_CalculosCOS')

df2 = df21.merge(df22, how='inner', on='CodParcela') 
```

```python
from pyproj import Proj, transform

proj_utm = {zone: Proj(proj='utm', zone=zone, ellps='WGS84', datum='WGS84') for zone in df2['Huso_C'].unique()}
proj_latlon = Proj(proj='latlong', datum='WGS84')

def convert_utm_to_latlon(row):
    utm_proj = proj_utm[row['Huso_C']]
    lon, lat = transform(utm_proj, proj_latlon, row['XUTM_ETRS89_C'], row['YUTM_ETRS89_C'])
    return pd.Series([lat, lon])

# Apply the conversion for each row
df2[['lat', 'lon']] = df2.apply(convert_utm_to_latlon, axis=1)
```

```python
data = df2

import matplotlib.colors as mcolors
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11, 8))
hexbin = ax.hexbin(data['lon'], data['lat'], gridsize=50, cmap='RdYlGn', mincnt=1) #, norm=mcolors.LogNorm()
ax.set_xlabel('Longitude', fontsize=14)
ax.set_ylabel('Latitude', fontsize=14)
ax.set_title(f'data distribution', fontsize=16)

ax.set_ylim([20,72])
ax.set_xlim([-20,35])
colorbar = plt.colorbar(hexbin)
colorbar.set_label(f'count', fontsize=14)

plt.grid(True)
plt.show()
```

```python
temp1 = pd.DataFrame()
temp1['time'] = df2['Fecha_C'].astype(str).str[0:4].astype(int)
temp1['lat'] = df2['lat']
temp1['lon'] = df2['lon']
temp1['hzn_top'] = 0
temp1['hzn_btm'] = 10
temp1['point_id'] = df2['CodParcela']

temp2 = pd.DataFrame()
temp2['time'] = df2['Fecha_C'].astype(str).str[0:4].astype(int)
temp2['lat'] = df2['lat']
temp2['lon'] = df2['lon']
temp2['hzn_top'] = 10
temp2['hzn_btm'] = 30
temp2['point_id'] = df2['CodParcela']

temp1['oc'] = df2['C % M1']*13 # % -> g/kg, walkey-black method
temp2['oc'] = df2['C % M2']*13 # % -> g/kg, walkey-black method

temp1['bulk_density'] = df2['DA M1'] 
temp2['bulk_density'] = df2['DA M2']

temp = pd.concat([temp1,temp2])

temp['ref'] = 'spain.ParcelasCOS'
temp['nuts0'] = 'ES'
```

```python
temp.to_csv(f'{data_path}/data/spain.ParcelasCOS_harmonized_v1.csv',index=False)
temp.head(5)
```
